```python
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
X = np.array([[ 99,  -1],
       [ 98,  -1],
       [ 97,  -2],
       [101,   1],
       [102,   1],
       [103,   2]])
plt.plot(X[:,0],X[:,0],c = "r", marker = "o")
pca2 = PCA(n_components=2)
pca2.fit(X)
print(f"Variance :{pca2.explained_variance_ratio_}")

# transforming the data
xTrans = pca2.transform(X)
```

```python
# PCA in action
def getpairs(n, df):
    # Get n random pairs of columns in a dataset
    from random import randint
    tuples = []
    i = 0
    while i < n:
        x = df.columns[randint(0,999)]
        y = df.columns[randint(0,999)]
        while x == y or (x,y) in tuples or (y,x) in tuples:
            y = df.columns[randint(0,999)]
        tuples.append((x,y))

        i+=1
    return tuples

df = pd.read_csv("./data/toy_dataset.csv")
tuples = getpairs(100, df)
```

```python
fig, ax = plt.subplots(10,10, figsize = (35,35))
axeFlat = ax.flatten()
for i in range(len(axeFlat)):
    axeFlat[i].scatter(df[tuples[i][0]],df[tuples[i][1]])
```

```python
pca2 = PCA(n_components=2)
xPca = pca2.fit(df)
xPca = pca2.transform(df) # You clould use fittransform method instead of last two lines
dfPCA2 = pd.DataFrame(xPca, columns=["PrincipalComponent1", "PrincipalComponent2"])
plt.scatter(dfPCA2.iloc[:,0], dfPCA2.iloc[:,1])
plt.xlabel("principal component 1")
plt.xlabel("principal component 2")
plt.show()

print(f"Preserved variancd: {sum(pca2.explained_variance_ratio_)}")
```

```python
pca3 = PCA(n_components=3)
xPca = pca3.fit(df)
xPca = pca3.transform(df) # You clould use fittransform method instead of last two lines
dfPCA3 = pd.DataFrame(xPca, columns=["PrincipalComponent1", "PrincipalComponent2", "PrincipalComponent3"])

fig = plt.figure()
ax = fig.add_subplot(projection="3d")
plt.scatter(dfPCA3.iloc[:,0], dfPCA3.iloc[:,1], dfPCA3.iloc[:,2])
plt.xlabel("principal component 1")
plt.xlabel("principal component 2")
plt.xlabel("principal component 3")
plt.show()

print(f"Preserved variancd: {sum(pca3.explained_variance_ratio_)}")
```
